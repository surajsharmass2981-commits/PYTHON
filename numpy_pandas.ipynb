## Import Library and Load Dataset

**Answer:**

Below code loads `Retail_Dataset.csv`.  
If the file is not available, it creates a small sample dataset so the notebook can still run.


```python
import pandas as pd
import numpy as np

try:
    df = pd.read_csv("Retail_Dataset.csv")
except FileNotFoundError:
    data = {
        "Date": ["2023-01-05", "2023-01-12", "2023-02-03", "2023-02-18", "2023-03-07",
                 "2023-03-15", "2023-04-01", "2023-04-20", "2023-05-09", "2023-05-25"],
        "Product Category": ["Electronics", "Clothing", "Grocery", "Electronics", np.nan,
                             "Clothing", "Furniture", "Grocery", "Electronics", "Furniture"],
        "Region": ["East", "West", "North", "South", "East",
                   np.nan, "West", "South", "North", "East"],
        "Quantity": [5, 10, np.nan, 7, 6, 8, 4, np.nan, 9, 3],
        "Price": [200.0, 50.0, 20.0, np.nan, 150.0, 80.0, 300.0, 25.0, np.nan, 400.0]
    }
    df = pd.DataFrame(data)

df
```

## Question 2

**Q2. Inspect the dataset and identify missing values in each column. Display the total number of missing values present in the dataset.**

**Answer:**

```python
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)

print("\nTotal missing values in the dataset:")
print(missing_values.sum())
```

## Question 3

**Q3. Replace missing values in the Quantity and Price columns with their respective mean values using Pandas.**

**Answer:**

```python
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].mean())
df["Price"] = df["Price"].fillna(df["Price"].mean())

df[["Quantity", "Price"]]
```

## Question 4

**Q4. Remove all rows where the Product Category or Region column contains missing values.**

**Answer:**

```python
df = df.dropna(subset=["Product Category", "Region"])
df
```

## Question 5

**Q5. Create a new column called Revenue using NumPy, where:

Revenue = Quantity × Price

Display the updated dataset with the new column.**

**Answer:**

```python
df["Revenue"] = np.multiply(df["Quantity"], df["Price"])
df
```

## Question 6

**Q6. Using NumPy, calculate the total revenue generated from all transactions in the dataset.**

**Answer:**

```python
total_revenue = np.sum(df["Revenue"])
print("Total Revenue:", total_revenue)
```

## Question 7

**Q7. Group the dataset by Product Category and calculate the total revenue generated by each category.**

**Answer:**

```python
category_revenue = df.groupby("Product Category")["Revenue"].sum()
category_revenue
```

## Question 8

**Q8. Based on the revenue calculated for each product category, identify the top 3 categories with the highest revenue and the bottom 3 categories with the lowest revenue.**

**Answer:**

```python
top_3_categories = category_revenue.sort_values(ascending=False).head(3)
bottom_3_categories = category_revenue.sort_values(ascending=True).head(3)

print("Top 3 Categories with Highest Revenue:")
print(top_3_categories)

print("\nBottom 3 Categories with Lowest Revenue:")
print(bottom_3_categories)
```

## Question 9

**Q9. Group the dataset by Region and calculate the total revenue generated in each region. Identify the region with the highest revenue and the region with the lowest revenue.**

**Answer:**

```python
region_revenue = df.groupby("Region")["Revenue"].sum()

print("Revenue by Region:")
print(region_revenue)

print("\nRegion with Highest Revenue:")
print(region_revenue.idxmax(), "-", region_revenue.max())

print("\nRegion with Lowest Revenue:")
print(region_revenue.idxmin(), "-", region_revenue.min())
```

## Question 10

**Q10. Extract the month from the Date column using Pandas and group the dataset by month to calculate the total monthly revenue. Also, calculate the mean, median, and standard deviation of the Revenue column using NumPy.**

**Answer:**

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month

monthly_revenue = df.groupby("Month")["Revenue"].sum()

print("Monthly Revenue:")
print(monthly_revenue)

print("\nMean Revenue:", np.mean(df["Revenue"]))
print("Median Revenue:", np.median(df["Revenue"]))
print("Standard Deviation of Revenue:", np.std(df["Revenue"]))
```

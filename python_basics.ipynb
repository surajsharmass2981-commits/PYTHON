## Question 2 : Explain the difference between List and Tuple in Python.

### Answer:

| List | Tuple |
|------|-------|
| List is mutable. | Tuple is immutable. |
| List uses square brackets `[]`. | Tuple uses round brackets `()`. |
| We can change elements in a list. | We cannot change elements in a tuple after creation. |
| Example: `[1, 2, 3]` | Example: `(1, 2, 3)` |

The main difference is that a **list can be modified**, but a **tuple cannot be modified**.

## Question 5 : Write a Python program to check whether a number is even or odd using conditional statements.

### Answer:

```python
number = int(input("Enter a number: "))

if number % 2 == 0:
    print("The number is even.")
else:
    print("The number is odd.")
```

## Question 8 : Write a Python program to remove duplicate values from a list using a set.

### Answer:

```python
numbers = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = list(set(numbers))
print("List after removing duplicates:", unique_numbers)
```

## Question 9 : Write a function that returns the square of a number.

### Answer:

```python
def square(num):
    return num * num

print(square(5))
```

## Question 10 : Write a Python program to count how many times a number appears in a list.

Example List
[2,3,4,2,5,2]

### Answer:

```python
numbers = [2, 3, 4, 2, 5, 2]
target = 2

count = numbers.count(target)
print("The number", target, "appears", count, "times in the list.")
```

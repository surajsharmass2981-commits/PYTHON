Load the dataset

Perform basic checks using:
- `df.info()`
- `df.describe()`

Handle missing values (if any)

```python
import pandas as pd
import numpy as np
import os

file_path = "House Price India.csv"

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print("Dataset loaded successfully.")
else:
    print("Dataset file not found. Please upload the file.")
    df = pd.DataFrame()

print("\nFirst 5 rows:")
print(df.head())

print("\nDataset info:")
df.info()

print("\nDescriptive statistics:")
print(df.describe())

print("\nMissing values in each column:")
missing_values = df.isnull().sum()
print(missing_values)

print("\nTotal missing values:", missing_values.sum())

if missing_values.sum() == 0:
    print("No missing values found.")
else:
    print("Missing values are present.")
```

Q1. What is the average, median, and standard deviation of house prices?

```python
average_price = df["Price"].mean()
median_price = df["Price"].median()
std_price = df["Price"].std()

print("Average house price:", round(average_price, 2))
print("Median house price:", round(median_price, 2))
print("Standard deviation of house price:", round(std_price, 2))

print("\nInterpretation:")
print("The average house price is", round(average_price, 2), ", the median is", round(median_price, 2), "and the standard deviation is", round(std_price, 2), ".")
```

Q2. Which number of bedrooms is most common?

```python
most_common_bedrooms = df["number of bedrooms"].mode()[0]

print("Most common number of bedrooms:", most_common_bedrooms)

print("\nInterpretation:")
print("The most common house type in this dataset has", most_common_bedrooms, "bedrooms.")
```

Q3. Check if the price data is skewed.

```python
price_skewness = df["Price"].skew()

print("Skewness of price data:", round(price_skewness, 4))

print("\nInterpretation:")
if price_skewness > 0:
    print("The price data is positively skewed.")
elif price_skewness < 0:
    print("The price data is negatively skewed.")
else:
    print("The price data is approximately symmetric.")
```

Q4. What is the average price for each number of bedrooms?

```python
average_price_by_bedrooms = df.groupby("number of bedrooms")["Price"].mean().sort_index()

print("Average price for each number of bedrooms:")
print(average_price_by_bedrooms.round(2))

print("\nInterpretation:")
print("In general, the average house price increases as the number of bedrooms increases.")
```

Q5. What is the relationship between living area and price? (Use : Area without Basement).

```python
area_col = "Area of the house(excluding basement)"
correlation_value = df[area_col].corr(df["Price"])

print("Correlation between area without basement and price:", round(correlation_value, 4))

print("\nInterpretation:")
if correlation_value > 0:
    print("There is a positive relationship between living area and price.")
elif correlation_value < 0:
    print("There is a negative relationship between living area and price.")
else:
    print("There is no linear relationship between living area and price.")
```

Q6. Identify any anomalies where houses have high prices but low areas.

```python
area_col = "Area of the house(excluding basement)"
price_threshold = df["Price"].quantile(0.85)
area_threshold = df[area_col].quantile(0.15)

anomalies = df[(df["Price"] > price_threshold) & (df[area_col] < area_threshold)][["id", "Price", area_col, "number of bedrooms", "Postal Code"]]

print("Price threshold used:", price_threshold)
print("Area threshold used:", area_threshold)
print("Number of anomalies found:", len(anomalies))

print("\nAnomalies:")
print(anomalies.head(10).to_string(index=False))

print("\nInterpretation:")
print("These houses have higher prices even though their area is relatively low.")
```

Q7. Compare average price based on number of floors and houses with or without waterfront.

```python
average_price_comparison = df.pivot_table(
    values="Price",
    index="number of floors",
    columns="waterfront present",
    aggfunc="mean"
)

print("Average price based on number of floors and waterfront:")
print(average_price_comparison.round(2))

print("\nInterpretation:")
print("Houses with waterfront usually have higher average prices than houses without waterfront.")
```

Q8. Identify the minimum and maximum house price. What does this indicate?

```python
minimum_price = df["Price"].min()
maximum_price = df["Price"].max()

print("Minimum house price:", minimum_price)
print("Maximum house price:", maximum_price)

print("\nInterpretation:")
print("This indicates that the dataset has a wide price range and includes both lower-priced and luxury houses.")
```

Q9. Which location (zipcode/area) has the highest average price?

```python
location_col = "zipcode" if "zipcode" in df.columns else "Postal Code"
average_price_by_location = df.groupby(location_col)["Price"].mean().sort_values(ascending=False)

print("Top locations by average price:")
print(average_price_by_location.head().round(2))

print("\nLocation with highest average price:", average_price_by_location.idxmax())
print("Highest average price:", round(average_price_by_location.max(), 2))

print("\nInterpretation:")
print("The location shown above has the highest average house price in the dataset.")
```

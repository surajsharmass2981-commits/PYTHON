```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

file_name = "employee_productivity_dataset.csv"

try:
    df = pd.read_csv(file_name)
except FileNotFoundError:
    print(f"{file_name} file not found.")
```

Q1. Handle Missing Values (Basic Cleaning)

Fill missing values using the following:

- Age → Median
- Salary → Mean
- Hours_Worked_Per_Week → Median
- Performance_Score → Mean

Display the dataset after handling missing values.

```python
q1_df = df.copy()

q1_df["Age"] = q1_df["Age"].fillna(q1_df["Age"].median())
q1_df["Salary"] = q1_df["Salary"].fillna(q1_df["Salary"].mean())
q1_df["Hours_Worked_Per_Week"] = q1_df["Hours_Worked_Per_Week"].fillna(q1_df["Hours_Worked_Per_Week"].median())
q1_df["Performance_Score"] = q1_df["Performance_Score"].fillna(q1_df["Performance_Score"].mean())

q1_df
```

Q2. Label Encoding

Convert these columns into numeric values using Label Encoding:

- Gender
- Department

Show the updated columns.

```python
q2_df = q1_df.copy()

gender_encoder = LabelEncoder()
department_encoder = LabelEncoder()

q2_df["Gender"] = gender_encoder.fit_transform(q2_df["Gender"])
q2_df["Department"] = department_encoder.fit_transform(q2_df["Department"])

q2_df[["Gender", "Department"]]
```

Q3. One-Hot Encoding

Apply One-Hot Encoding on:

- Work_Mode
- Location

Display the dataset and check how many new columns are created.

```python
q3_df = pd.get_dummies(q2_df, columns=["Work_Mode", "Location"])

new_columns = [col for col in q3_df.columns if col.startswith("Work_Mode_") or col.startswith("Location_")]

print("New columns created:", len(new_columns))
print(new_columns)

q3_df
```

Q4. Normalization (Min-Max Scaling)

Normalize the following columns:

- Salary
- Hours_Worked_Per_Week

Use MinMaxScaler and display results.

```python
q4_df = q3_df.copy()

minmax_scaler = MinMaxScaler()
q4_df[["Salary", "Hours_Worked_Per_Week"]] = minmax_scaler.fit_transform(
    q4_df[["Salary", "Hours_Worked_Per_Week"]]
)

q4_df[["Salary", "Hours_Worked_Per_Week"]]
```

Q5. Standardization (Scaling)

Apply StandardScaler on:

- Age
- Projects_Completed

Display the transformed values.

```python
q5_df = q4_df.copy()

standard_scaler = StandardScaler()
q5_df[["Age", "Projects_Completed"]] = standard_scaler.fit_transform(
    q5_df[["Age", "Projects_Completed"]]
)

q5_df[["Age", "Projects_Completed"]]
```

Q6. Compare Scaling Methods

Apply both:

- MinMaxScaler
- StandardScaler

On the Salary column.

Show both results side by side.

```python
salary_compare = q1_df[["Salary"]].copy()

salary_minmax = MinMaxScaler()
salary_standard = StandardScaler()

compare_df = pd.DataFrame({
    "Salary_MinMaxScaler": salary_minmax.fit_transform(salary_compare).flatten(),
    "Salary_StandardScaler": salary_standard.fit_transform(salary_compare).flatten()
})

compare_df
```

Q7. Build Preprocessing Pipeline

Create a pipeline that:

- Applies encoding to categorical columns
- Applies scaling to numerical columns

Use:

- ColumnTransformer
- Pipeline

```python
categorical_columns = ["Gender", "Department", "Work_Mode", "Location"]
numerical_columns = ["Age", "Salary", "Hours_Worked_Per_Week", "Projects_Completed", "Performance_Score"]

categorical_pipeline = Pipeline(steps=[
    ("encoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

numerical_pipeline = Pipeline(steps=[
    ("scaler", StandardScaler())
])

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", categorical_pipeline, categorical_columns),
        ("num", numerical_pipeline, numerical_columns)
    ]
)

pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor)
])

pipeline
```

Q8. Apply Pipeline

Apply the pipeline on the dataset.

Display:

- Transformed dataset
- Shape of final dataset

```python
pipeline_data = q1_df[categorical_columns + numerical_columns].copy()

transformed_data = pipeline.fit_transform(pipeline_data)
feature_names = pipeline.named_steps["preprocessor"].get_feature_names_out()

transformed_df = pd.DataFrame(transformed_data, columns=feature_names)

print("Shape of final dataset:", transformed_df.shape)

transformed_df
```

```python
import os
import pandas as pd

file_path = "Advanced_pandas_sales_dataset.csv"

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print("Dataset loaded successfully.")
else:
    print("Dataset file is missing.")
```

Q1 Dataset Exploration

Perform the following exploratory steps:

- Display the first 10 rows
- Display the dataset structure using info()
- Find the number of rows and columns
- Identify the data types of each column


```python
print("First 10 rows:")
df.head(10)
print("\nDataset structure:")
df.info()
print("\nNumber of rows and columns:")
print(df.shape)
print("\nData types of each column:")
df.dtypes
```

Q2 Sales Filtering Analysis

Filter the dataset to display:

1. Orders where Sales > 3500
2. Orders where Region = "East" and Sales > 2500

Explain what insights you observe.


```python
print("Orders where Sales > 3500:")
sales_above_3500 = df[df["Sales"] > 3500]
sales_above_3500

print("\nOrders where Region = 'East' and Sales > 2500:")
east_sales_above_2500 = df[(df["Region"] == "East") & (df["Sales"] > 2500)]
east_sales_above_2500

print("\nOutput:")
print("There are many high-value orders in the dataset.")
print("East region also has many orders above 2500.")
print("Some high-sales orders have low profit or negative profit.")
```

Q3 Customer Purchase Analysis

Identify the top 5 customers based on total sales.

Hint: Use GroupBy with aggregation.


```python
top_5_customers = df.groupby("Customer")["Sales"].sum().reset_index()
top_5_customers = top_5_customers.sort_values("Sales", ascending=False).head(5)
top_5_customers
```

Q4 Regional Sales Performance

Calculate the total sales generated by each region.

Then identify:

- Which region generated the highest revenue?


```python
region_sales = df.groupby("Region")["Sales"].sum().reset_index()
region_sales = region_sales.sort_values("Sales", ascending=False)
print("Total sales generated by each region:")
region_sales

print("\nOutput:")
print("Region with highest revenue:", region_sales.iloc[0]["Region"])
```

Q5 Product Performance Analysis

- Calculate the total quantity sold for each product.
- Which product appears to be the most popular?


```python
product_quantity = df.groupby("Product")["Quantity"].sum().reset_index()
product_quantity = product_quantity.sort_values("Quantity", ascending=False)
print("Total quantity sold for each product:")
product_quantity

print("\nOutput:")
print("Most popular product:", product_quantity.iloc[0]["Product"])
```

Q6 Multi-Level Grouping

- Group the dataset by: Region and Category
- Calculate the total sales for each combination.
- What insights can you derive from this analysis?


```python
region_category_sales = df.groupby(["Region", "Category"])["Sales"].sum().reset_index()
region_category_sales = region_category_sales.sort_values(["Region", "Sales"], ascending=[True, False])
print("Total sales for each Region and Category combination:")
region_category_sales

highest_combo = region_category_sales.sort_values("Sales", ascending=False).iloc[0]
lowest_combo = region_category_sales.sort_values("Sales", ascending=True).iloc[0]

print("\nOutput:")
print(f"Highest sales combination: {highest_combo['Region']} - {highest_combo['Category']}")
print(f"Lowest sales combination: {lowest_combo['Region']} - {lowest_combo['Category']}")
print("Electronics category performs better than other categories in most regions.")
```

Q7 Pivot Table Creation

Create a pivot table showing:

- Rows -> Region
- Columns -> Product
- Values -> Total Sales

Which region sells the most laptops?


```python
pivot_table_sales = pd.pivot_table(df, index="Region", columns="Product", values="Sales", aggfunc="sum", fill_value=0)
print("Pivot table:")
pivot_table_sales

print("\nOutput:")
print("Region that sells the most laptops:", pivot_table_sales["Laptop"].idxmax())
```

Q8 Profitability Analysis

- Calculate the average profit for each product category.
- Which category is most profitable?


```python
avg_profit_category = df.groupby("Category")["Profit"].mean().reset_index()
avg_profit_category = avg_profit_category.sort_values("Profit", ascending=False)
print("Average profit for each product category:")
avg_profit_category

print("\nOutput:")
print("Most profitable category:", avg_profit_category.iloc[0]["Category"])
```

Q9 Discount Impact Study

Analyze the relationship between discount and profit.

Tasks:

- Calculate the average profit for each discount level
- Identify whether higher discounts lead to lower profit


```python
discount_profit = df.groupby("Discount")["Profit"].mean().reset_index()
discount_profit = discount_profit.sort_values("Discount")
print("Average profit for each discount level:")
discount_profit

print("\nOutput:")
print("Yes, higher discounts lead to lower average profit in this dataset.")
```

Q10 End-to-End Data Wrangling Task

Perform the following workflow:

1. Filter the dataset where Sales > 3000
2. Group data by Region and Product
3. Calculate total sales
4. Reshape the result into a pivot table

Explain the business insights obtained from this analysis.


```python

filtered_df = df[df["Sales"] > 3000]
grouped_sales = filtered_df.groupby(["Region", "Product"])["Sales"].sum().reset_index()
pivot_result = grouped_sales.pivot(index="Region", columns="Product", values="Sales").fillna(0)

print("Grouped data:")
grouped_sales

print("\nPivot table:")
pivot_result

print("\nOutput:")
print("East region shows strong sales in Laptop, Phone and Tablet.")
print("West region is strong in Keyboard.")
print("South region is strong in Mouse.")
print("High-value sales are spread across different products and regions.")
```

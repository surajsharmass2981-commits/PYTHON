Q1. Creating a DataFrame

Create a Pandas DataFrame using a Python dictionary with the following columns:

Product

Region

Sales

Quantity

Add at least 5 rows of sample data and display the DataFrame.

```python
import pandas as pd

data = {
    "Product": ["A","B","C","D","E"],
    "Region": ["East","West","North","South","East"],
    "Sales": [1000,1500,2000,1200,1300],
    "Quantity": [10,15,20,12,13]
}

df = pd.DataFrame(data)
df
```

Q4. Indexing using .loc

Q9. Descriptive Statistics

Generate summary statistics using:

describe()

From the result identify:

Mean sales value

Maximum quantity sold

Minimum sales value

```python
desc = df.describe()
print(desc)
print("Mean sales value:", desc.loc['mean','Sales'])
print("Maximum quantity sold:", desc.loc['max','Quantity'])
print("Minimum sales value:", desc.loc['min','Sales'])
```

Q10. Aggregation Analysis

Using the Sales column, calculate:

Total sales

Average sales

Maximum sales

Minimum sales

```python
print("Total sales:", df['Sales'].sum())
print("Average sales:", df['Sales'].mean())
print("Maximum sales:", df['Sales'].max())
print("Minimum sales:", df['Sales'].min())
```
